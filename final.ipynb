```python
#Kütüphane Kataloglama sistemi

from google.colab import drive
drive.mount('/content/drive')
dosya = "/content/drive/MyDrive/Colab Notebooks/veri.txt"


print("Kütüphane Kataloglama Sistemi \n")
print ("Hoş geldin!")
def kitapListele():
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()
  menu()

def kitapKaydet():
  kitapAdi = input("Kitabın adını giriniz: ")
  kitapYazari = input("Kitabın yazarını giriniz: ")
  kitapBasimTarihi = input("Kitabın basım tarihini giriniz: ")
  f = open(dosya, 'a' )
  f.write(kitapAdi + " - " + kitapYazari + " - " + " - " + kitapBasimTarihi + " - " +'\n')
  print("Kitap kaydını tamamladın")
  f.close()
  menu()

def kitapAra():
  kayitAra = input("Arama yapmak istediğiniz kitabın adını girin: ")
  kayit = open(dosya, "r")

  arananKayit = kayit.read().find(kayitAra)
  if arananKayit != -1:
    print("Aradığınız kitap mevcuttur.")
    menu()
  else:
    print("Aradığınız kitap katalogta yok.")
    menu()

def sil():
    silinecekKitap = input("Silmek istediğin kitabın adını gir: ")
    with open(dosya, 'r') as f:
        lines = f.readlines()

    with open(dosya, 'w') as f:
        for line in lines:
            if silinecekKitap not in line:
                f.write(line)

    print("Kitap silindi.")
    menu()

def katalogTemizle():
  onay = input("Katalogu tamamen temizlemek istediğinizden emin misiniz? (E/H): ")
  if onay.lower() == "e":
    with open(dosya, 'w') as f:
      f.write("")
    print("Katalog tamamen temizlendi.")
  else:
    print("İşlem iptal edildi.")

  menu()


def cikis():
  print("\t\t\n Katalogtan çıkış yapmak istiyor musun?\n \t1) Evet\n  \t2) Hayır\n\t\t")
  cikisIslem = input("İşlem seç: ")
  if cikisIslem == "2":
    print("Menüye yönlendiriliyorsun. Seçimini yap. ")
    menu()
  elif cikisIslem == "1":
   print("Çıkış yapıldı.")
  else:
    print("Yanlış işlem. Sadece 1 ve 2 seçebilirsin. ")
    cikis()

def menu():
  print("\t\t\n Yapmak istediğin işlemi seç:\n  \t1) Kitap Listele\n  \t2) Kitap Kaydet\n  \t3) Kitap Ara\n  \t4) Sil\n \t5) Katalogu Tamamen Temizle\n \t6) Çıkış Yap\n\t\t")
  islem = input("İşlem seç: ")
  if islem == "1":
    kitapListele()
  elif islem == "2":
    kitapKaydet()
  elif islem== "3":
    kitapAra()
  elif islem == "4":
    sil()
  elif islem == "5":
    katalogTemizle()
  elif islem == "6":
    cikis()
    
  else:
    print("Yanlış giriş yaptın! Menüye yönlendiriliyorsun.")
    menu()

menu()
```
